## Splitting the data between train and test

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size = 0.2,random_state = 42)
```

## Skewness

```python
train_set.skew()
```

```python
train_set['generation_gwh_2013'] = np.log1p(train_set['generation_gwh_2013'])
train_set['generation_gwh_2014'] = np.log1p(train_set['generation_gwh_2014'])
train_set['generation_gwh_2015'] = np.log1p(train_set['generation_gwh_2015'])
train_set['generation_gwh_2016'] = np.log1p(train_set['generation_gwh_2016'])
train_set['generation_gwh_2017'] = np.log1p(train_set['generation_gwh_2017'])
```

```python
test_set['generation_gwh_2013'] = np.log1p(test_set['generation_gwh_2013'])
test_set['generation_gwh_2014'] = np.log1p(test_set['generation_gwh_2014'])
test_set['generation_gwh_2015'] = np.log1p(test_set['generation_gwh_2015'])
test_set['generation_gwh_2016'] = np.log1p(test_set['generation_gwh_2016'])
test_set['generation_gwh_2017'] = np.log1p(test_set['generation_gwh_2017'])
```

```python
train_set.skew()
```

 ## Splitting the data into X_train, X_test, y_train, y_test for Capacity in Megawatt prediction.

```python
from sklearn.model_selection import train_test_split

X_train = train_set.drop('capacity_mw', axis=1)
y_train = train_set['capacity_mw']

X_test = test_set.drop('capacity_mw', axis=1)
y_test = test_set['capacity_mw']
```

## Scaling the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=16)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
explained_variance_ratio_ = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio_)

plt.plot(range(1, len(explained_variance_ratio_) + 1), explained_variance_ratio_, marker='o', linestyle='--')
plt.plot(range(1, len(explained_variance_ratio_) + 1), cumulative_explained_variance, marker='o', linestyle='-')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```

```python
sum(pca.explained_variance_ratio_)
```

## Models

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

lr = LinearRegression()
lr.fit(X_train_pca, y_train)

y_pred = lr.predict(X_test_pca)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=5, include_bias=False)
X_poly_train = poly_features.fit_transform(X_train_pca)
X_poly_test = poly_features.transform(X_test_pca)

poly_reg = LinearRegression()
poly_reg.fit(X_poly_train, y_train)

y_poly_pred = poly_reg.predict(X_poly_test)

mae = mean_absolute_error(y_test, y_poly_pred)
mse = mean_squared_error(y_test, y_poly_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_poly_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(max_iter=1000, tol=1e-3, penalty=None, eta0=0.1)
sgd_reg.fit(X_train_pca, y_train)

y_pred = sgd_reg.predict(X_test_pca)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators = 300, max_depth = 100,
                            min_samples_split = 2, random_state=42)
rfr.fit(X_train_pca, y_train)

y_pred = rfr.predict(X_test_pca)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

The best model for Regression is Random Forest Regressor with a MAE of 124.
