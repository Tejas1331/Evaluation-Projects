## Importing libraries and dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/temperature.csv')
```

```python
df
```

## Creating a Copy of the dataframe

```python
df_copy_1 = df.copy()
```

## Checking for Null values

```python
df.info()
```

## Data Filtering and Understanding for every column

Column1 - Station : It is the station number from where the data is taken and the value ranges from 1 to 25.

```python
df['station'].unique()
```

```python
df['station'].isna().sum()
```

```python
df['station'].value_counts()
```

Comments - We will convert the station values to Integer type and fill the nan values with median of the station.

```python
df['station'] = df['station'].fillna(df['station'].median())
```

```python
df['station'] = df['station'].astype(int)
```

Column2 - Date : Date can play as a very important component as temperature and seasons change month wise.

Coverting to datetime

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date'].value_counts()
```

```python
df['Date'].isna().sum()
```

Comments - Logically speaking from the data every month has been alloted 25 temperature values, hence the 9th month in 2017 will be used to replace the 2 missing values. But since 2 values won't influence our data that much considering the large size of the data. It is ok to get rid of these values.

```python
df = df.dropna(subset=['Date'])
```

Now we will fetch the months out the the date and assign the integer values to the months as we want the model to understand that there is an ordinal relationship between the month and the temperature, as we all know temperatures change with season.

```python
df['Date'][0]
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df['Month'] = df['Month'].astype(int)
```

```python
df['Month'].unique()
```

```python
df['Day'] = df['Date'].dt.day
```

```python
df['Day'] = df['Day'].astype(int)
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df
```

Now we will drop the date column

```python
df = df.drop(['Date'], axis=1)
```

```python
df.info()
```

## Deleting the rows which have empty Next_Tmax and Next_Tmin as they are our predictor variables

```python
df = df.dropna(subset=['Next_Tmax', 'Next_Tmin'])
```

```python
df.info()
```

## Filling the empty or null values

```python
df['Present_Tmax'].fillna(df['Present_Tmax'].mean(), inplace=True)
df['Present_Tmin'].fillna(df['Present_Tmin'].mean(), inplace=True)
```

```python
ldaps_columns = [
    'LDAPS_RHmin', 'LDAPS_RHmax', 'LDAPS_Tmax_lapse', 'LDAPS_Tmin_lapse',
    'LDAPS_WS', 'LDAPS_LH', 'LDAPS_CC1', 'LDAPS_CC2', 'LDAPS_CC3', 'LDAPS_CC4',
    'LDAPS_PPT1', 'LDAPS_PPT2', 'LDAPS_PPT3', 'LDAPS_PPT4'
]

df[ldaps_columns] = df[ldaps_columns].interpolate()
```

```python
df.info()
```

## Setting aside Train and Test set

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

## Skewness

```python
train_set.skew()
```

Seems like LDAPS_PPT1, LDAPS_PPT2, LDAPS_PPT3, LDAPS_PPT4 are extremely skewed.We will get rid of the skewness.

```python
train_set['LDAPS_PPT1'] = np.power(train_set['LDAPS_PPT1'], 1/3)
train_set['LDAPS_PPT2'] = np.power(train_set['LDAPS_PPT2'], 1/3)
train_set['LDAPS_PPT3'] = np.power(train_set['LDAPS_PPT3'], 1/3)
train_set['LDAPS_PPT4'] = np.power(train_set['LDAPS_PPT4'], 1/3)
```

```python
train_set.skew()
```

```python
test_set['LDAPS_PPT1'] = np.power(test_set['LDAPS_PPT1'], 1/3)
test_set['LDAPS_PPT2'] = np.power(test_set['LDAPS_PPT2'], 1/3)
test_set['LDAPS_PPT3'] = np.power(test_set['LDAPS_PPT3'], 1/3)
test_set['LDAPS_PPT4'] = np.power(test_set['LDAPS_PPT4'], 1/3)
```

## Spiltting into X_train, y_train and X_test, y_test for NEXT_Tmax

```python
X_train_Tmax = train_set.drop(['Next_Tmax'], axis=1)
y_train_Tmax = train_set['Next_Tmax']
```

```python
X_test_Tmax = test_set.drop(['Next_Tmax'], axis=1)
y_test_Tmax = test_set['Next_Tmax']
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_Tmax_scaled = scaler.fit_transform(X_train_Tmax)
X_test_Tmax_scaled = scaler.transform(X_test_Tmax)
```

## PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
X_train_Tmax_pca = pca.fit_transform(X_train_Tmax_scaled)
X_test_Tmax_pca = pca.transform(X_test_Tmax_scaled)
```

```python
explained_variance_ratio_ = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio_)

plt.plot(range(1, len(explained_variance_ratio_) + 1), explained_variance_ratio_, marker='o', linestyle='--')
plt.plot(range(1, len(explained_variance_ratio_) + 1), cumulative_explained_variance, marker='o', linestyle='-')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```

## Models

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

lr = LinearRegression()
lr.fit(X_train_Tmax_pca, y_train_Tmax)

y_pred = lr.predict(X_test_Tmax_pca)

mae = mean_absolute_error(y_test_Tmax, y_pred)
mse = mean_squared_error(y_test_Tmax, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_Tmax, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=3, include_bias=False)
X_poly_train = poly_features.fit_transform(X_train_Tmax_pca)
X_poly_test = poly_features.transform(X_test_Tmax_pca)

poly_reg = LinearRegression()
poly_reg.fit(X_poly_train, y_train_Tmax)

y_poly_pred = poly_reg.predict(X_poly_test)

mae = mean_absolute_error(y_test_Tmax, y_poly_pred)
mse = mean_squared_error(y_test_Tmax, y_poly_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_Tmax, y_poly_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(max_iter=1000, tol=1e-3, penalty=None, eta0=0.1)
sgd_reg.fit(X_train_Tmax_pca, y_train_Tmax)

y_pred = sgd_reg.predict(X_test_Tmax_pca)

mae = mean_absolute_error(y_test_Tmax, y_pred)
mse = mean_squared_error(y_test_Tmax, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_Tmax, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth = 4, random_state=42)
tree_reg.fit(X_train_Tmax_pca, y_train_Tmax)

y_pred = tree_reg.predict(X_test_Tmax_pca)

mae = mean_absolute_error(y_test_Tmax, y_pred)
mse = mean_squared_error(y_test_Tmax, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_Tmax, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators = 500, max_depth = 100,
                            min_samples_split = 2, random_state=42)
rfr.fit(X_train_Tmax_pca, y_train_Tmax)

y_pred = rfr.predict(X_test_Tmax_pca)

mae = mean_absolute_error(y_test_Tmax, y_pred)
mse = mean_squared_error(y_test_Tmax, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_Tmax, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbrt = GradientBoostingRegressor(max_depth = 2, n_estimators = 100, learning_rate = 0.1, random_state=42)
gbrt.fit(X_train_Tmax_pca, y_train_Tmax)

y_pred = gbrt.predict(X_test_Tmax_pca)

mae = mean_absolute_error(y_test_Tmax, y_pred)
mse = mean_squared_error(y_test_Tmax, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_Tmax, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

## Conclusion For Next_Tmax

The best model for prediction of Next_Tmax is Polynomial Regression with an MAE of :0.74.
